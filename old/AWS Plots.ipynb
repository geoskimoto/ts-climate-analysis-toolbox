```python
#Actual def that converts to water year called get_wy_date_series
#This file is the util.awdb.py file I do believe
import argparse
import calendar as cal
import csv
import datetime
import glob
import json
import logging
import math
import os
import urllib
import warnings
from concurrent.futures import as_completed
from logging.handlers import TimedRotatingFileHandler

import numpy as np
import pandas as pd
import plotly.graph_objs as go
import plotly.offline as py
from plotly.io import to_html as plotly_to_html
from requests_cache import CachedSession
from requests_futures.sessions import FuturesSession
# from utils.template import get_template, get_css

THIS_DIR = os.path.dirname(os.path.abspath(__file__))
SCRIPT_DIR = os.path.dirname(THIS_DIR)
ROOT_DIR = os.path.dirname(SCRIPT_DIR)
STATIC_DIR = os.path.join(ROOT_DIR, r"static")

dt = datetime.datetime
date = datetime.date
today = dt.utcnow() - datetime.timedelta(hours=8)
MAX_CACHE = 2000
JSON_WORKERS = 8

BASIN_NETWORKS = ["SNTL", "SNTLT"]
SITE_NETWORKS = ["SNTL", "SCAN", "SNTLT", "OTHER", "MSNT"]
DEFAULT_DOMAIN = "https://www.nrcs.usda.gov/Internet/WCIS"


SENSOR_LONG = {
    "WTEQ": "Snow Water Equivalent",
    "PREC": "Precipitation Accumulation",
    "TAVG": "Daily Average Temperature",
    "SMS": 'Avg. Soil Moisture (8",20")',
    "STO": 'Avg. Soil Temperature (2",4",8")',
    "SNWD": "Snow Depth",
}

SENSOR_UNITS = {
    "WTEQ": "in.",
    "PREC": "in.",
    "TAVG": "&#176;F",
    "SMS": "% Sat.",
    "STO": "&#176;F",
    "SNWD": "in.",
}

BASIN_STATES = [
    "AK",
    "AZ",
    "CA",
    "CO",
    "ID",
    "MT",
    "NM",
    "NV",
    "OR",
    "UT",
    "WA",
    "WY",
]

STATES_LONG = {
    "AK": "Alaska",
    "AL": "Alabama",
    "AR": "Arkansas",
    "AS": "American Samoa",
    "AZ": "Arizona",
    "CA": "Eastern Sierra",
    "CO": "Colorado",
    "CT": "Connecticut",
    "DC": "District of Columbia",
    "DE": "Delaware",
    "FL": "Florida",
    "GA": "Georgia",
    "GU": "Guam",
    "HI": "Hawaii",
    "IA": "Iowa",
    "ID": "Idaho",
    "IL": "Illinois",
    "IN": "Indiana",
    "KS": "Kansas",
    "KY": "Kentucky",
    "LA": "Louisiana",
    "MA": "Massachusetts",
    "MD": "Maryland",
    "ME": "Maine",
    "MI": "Michigan",
    "MN": "Minnesota",
    "MO": "Missouri",
    "MP": "Northern Mariana Islands",
    "MS": "Mississippi",
    "MT": "Montana",
    "NA": "National",
    "NC": "North Carolina",
    "ND": "North Dakota",
    "NE": "Nebraska",
    "NH": "New Hampshire",
    "NJ": "New Jersey",
    "NM": "New Mexico",
    "NV": "Nevada",
    "NY": "New York",
    "OH": "Ohio",
    "OK": "Oklahoma",
    "OR": "Oregon",
    "PA": "Pennsylvania",
    "PR": "Puerto Rico",
    "RI": "Rhode Island",
    "SC": "South Carolina",
    "SD": "South Dakota",
    "TN": "Tennessee",
    "TX": "Texas",
    "UT": "Utah",
    "VA": "Virginia",
    "VI": "Virgin Islands",
    "VT": "Vermont",
    "WA": "Washington",
    "WI": "Wisconsin",
    "WV": "West Virginia",
    "WY": "Wyoming",
}

PLOTLY_SRC = (
    "/assets/js/plotly.js"  # TODO: should probably make sure this is the min version
)

GA_JS = """
"""

GA_IFRAME = """
"""

BASIN_SITES_URL_DICT = {
    "prefix": "https://wcc.sc.egov.usda.gov/reportGenerator"
    "/view/customMultipleStationReport/daily/start_of_period/",
    "suffix": "id=%22%22%7Cname/0,0/stationId,state.code,"
    "network.code,name,elevation,latitude,longitude,"
    "county.name,huc8.huc,huc8.hucName,inServiceDate"
    ",outServiceDate?fitToScreen=false",
    "delim": "%7C",
    "text": "Station List",
}

ANNO_STATS = (
    "Statistical shading breaks at 10th, 30th, 50th, 70th, and 90th Percentiles"
)

NRCS_LOGO_SRC = "/assets/img/usda-logo.svg"

NORMS_LINK = "https://www.nrcs.usda.gov/resources/data-and-reports/climatic-and-hydrologic-normals"  # TODO: get drupal link

POR_ONLY_TXT = "*POR data used to calculate Normals since no published 30-year normals available for this site"


def get_basin_sites_link(valid_triplets, basin_sites_url_dict=BASIN_SITES_URL_DICT):
    site_list_triplets = [
        f"{triplet}{basin_sites_url_dict['delim']}" for triplet in valid_triplets
    ]
    sites_link = (
        f"<a href='{basin_sites_url_dict['prefix']}"
        f"{''.join(site_list_triplets)}{basin_sites_url_dict['suffix']}'>"
        f"{basin_sites_url_dict['text']}</a>"
    )
    return sites_link


def get_html_head_str(google_js=GA_JS):
    return f"""
    <head>
    {google_js}
    <!-- Created at: {datetime.datetime.now():%Y-%m-%d %H:%M} -->
    <title>AWS Plot</title>
    <link rel="shortcut icon" type="image/ico" href="/assets/favicon.ico">
    <link rel="icon" type="image/png" href="/assets/favicon.png">
"""


def get_html_body_str(plotly_src=PLOTLY_SRC, google_iframe=GA_IFRAME):
    return f"""
    <body>
    {google_iframe}
    <script src="{plotly_src}"></script>
"""


def get_no_chart_msg():
    return """
    <h4>No Chart Created!</h4>
    <p>This chart could not be created due to lack of data</p>
"""


def get_anno_str(anno_append="", anno_stats=ANNO_STATS, norms_link=NORMS_LINK):
    if anno_append == "no_chart":
        return get_no_chart_msg()
    anno_norm_link = (
        f"For more information visit: "
        f"<a href='{norms_link}'>"
        f"30-Year Hydroclimatic Normals</a>"
    )
    anno = f"{anno_stats}<br>{anno_norm_link}"
    if anno_append:
        anno = f"{anno}<br>{anno_append}"
    anno = f"{anno}<br>Updated: {dt.now():%A, %b %d, %Y %I %p CST}"
    return anno


def add_cdn(plotName, anno_append=""):
    with open(plotName, "r") as plotFile:
        plotStr = plotFile.read()
    plotStr = "<!DOCTYPE html>\n" + plotStr
    plotStr = plotStr.replace("<html>", '<html lang="en-US">')
    # plotStr = plotStr.replace("<body>", html_body_str)
    plotStr = plotStr.replace("<head>", get_html_head_str())
    plotStr = plotStr.replace("<body>", get_html_body_str())
    plotStr = plotStr.replace(
        "<div>", '<div style="height: 90vh; width: 98%; margin: auto;">'
    )
    plotStr = plotStr.replace("</body>", f"{get_anno_str(anno_append)}</body>")
    with open(plotName, "w") as newPlotFile:
        newPlotFile.write(plotStr)


def isActive(x):
    if dt.strptime(x.endDate, "%Y-%m-%d %H:%M:%S").date() > today.date():
        return True
    return False


def isAbove(x, elev):
    if x.elevation >= elev:
        return True
    return False


def isBelow(x, elev):
    if x.elevation <= elev:
        return True
    return False


def isInState(x, state):
    if x["stationTriplet"].split(":")[1].lower() == state:
        return True
    return False


def isYearsOld(x, yrs):
    s = str(x.beginDate)
    c = dt.today().year - yrs
    if int(s[:4]) < c:
        return True
    return False


def get_last_non_zero_index(d, default=366):
    rev = (len(d) - idx for idx, item in enumerate(reversed(d), 1) if item)
    return next(rev, default)


def ordinal(n):
    return "%d%s" % (
        n,
        "tsnrhtdd"[(math.floor(n // 10) % 10 != 1) * (n % 10 < 4) * n % 10 :: 4],
    )


def is_leap_wy(date_as_dt):
    year = date_as_dt.year
    if date_as_dt.month > 9:
        year += 1
    return is_leap_cal_yr(year)


def is_leap_cal_yr(year):
    return year % 4 == 0 and (year % 100 != 0 or year % 400 == 0)


def get_site_data_from_files(url_list, url_ids, args):
    requested_data = []
    for i, url_path in enumerate(url_list):
        file_path = os.path.join(*(url_path.split("/")))

        full_file_path = os.path.join(args.path, file_path).replace("/", os.sep)
        try:
            with open(full_file_path, "r") as file:
                file_content = file.read()
            json_content = json.loads(file_content)
        except Exception as exc:
            print(f"          Could not read file " f"{file_path} - {exc}")
        else:
            if isinstance(json_content, dict):
                json_content["id"] = url_ids[i]
            requested_data.append(json_content)
    return requested_data


def get_site_data(
    url_list, url_ids, args, domain=DEFAULT_DOMAIN, script_dir=SCRIPT_DIR
):  # max_workers=JSON_WORKERS
    cache_path = os.path.join(
        script_dir, "data", ".cache", f"{args.sensor.lower()}_json.sqlite"
    )
    # swap out to above line if using sqllite backend
    cache_path = args.sensor.lower()

    if args.data_source == "file":
        return get_site_data_from_files(url_list, url_ids, args)
    if args.url:
        domain = args.url
    success_data = []
    failed_data = []
    # bt = time.time()
    with FuturesSession(
        session=CachedSession(
            backend="memory",
            cache_name=cache_path,
            expire_after=60 * 60 * 2,
            allowable_codes=(200, 404),
        )
    ) as session:
        future_responses = []
        for i, url in enumerate(url_list):
            future = session.get(f"{domain}/{url}", timeout=30)
            future.id = url_ids[i]
            future.url = url
            future_responses.append(future)

        for future in as_completed(future_responses):
            try:
                resp = future.result()
                data = resp.json()
                if isinstance(data, dict):
                    data["id"] = future.id
                    data["url"] = resp.url
            except Exception:  # as e:
                failed_data.append(f"Failed to get data from {future.url}")  # - {e}")
            else:
                success_data.append(data)
    for fail in failed_data:
        print(fail)
    # print(f'{time.time() - bt} seconds')
    return success_data


# def url_to_json(url, url_id, sesh, domain=DEFAULT_DOMAIN):
#     url = domain + "/" + url
#     with sesh.get(url, timeout=30) as d:
#         data = d.json()
#         if isinstance(data, dict):
#             data['id'] = url_id
#     return data

# def get_site_data(url_list, url_ids, args, max_workers=JSON_WORKERS):
#     if args.data_source == "file":
#         return get_site_data_from_files(url_list, url_ids, args)

#     a persistent session and multi threaded json fetch
#     requested_data = []
#     for i, url in enumerate(url_list):
#         try:
#             data = url_to_json(url=url, url_id=url_ids[i], domain=DEFAULT_DOMAIN)
#         except Exception as e:
#             print(e)
#         else:
#             requested_data.append(data)
#     return requested_data
#     with CachedSession(backend='sqlite', expire_after=60*60*2) as sesh:
#         requested_data = []
#         with futures.ThreadPoolExecutor(max_workers=JSON_WORKERS) as executor:
#             future_to_url = {executor.submit(url_to_json, url, url_ids[i], sesh, args.url): url for
#                               i, url in enumerate(url_list)}
#             for i, future in enumerate(futures.as_completed(future_to_url)):
#                 url = future_to_url[future]
#                 try:
#                     data = future.result()
#                 except Exception as exc:
#                     if 'JSON' not in type(exc).__name__:
#                         print(f'          Could not download '
#                             f'{url.replace(args.url + "/" ,"")} - {exc}')
#                 else:
#                     requested_data.append(data)

#     For a non threaded solution use the persistent session below
#     with requests.Session() as sesh:
#         raw_data = [url_to_json(url, sesh, url_ids[i], args.url) for i, url in enumerate(url_list)]
#     requested_data = [d.json() for d in raw_data if d.status_code == 200]

#     errors = [f'Could not download {d.url}'
#         f'- {d.status_code} was returned' for d in raw_data
#         if d.status_code != 200]

#     if errors: print('\n'.join(errors))

#     return requested_data


def fillMissingData(x, daysBack):
    daysBack = -1 * daysBack
    if math.isnan(sum(x[daysBack:])) or not x:
        return x

    if len(x) < daysBack:
        daysBack = len(x)
    if math.isnan(x[-1]):
        x[-1] = [i for i in x if not math.isnan(i)][-1]
    y = x[daysBack:]
    x[:] = x[:daysBack] + pd.DataFrame(y).interpolate().values.ravel().tolist()
    return x


def nonLeapDaysBetween(_sDateLeap, _eDateLeap):
    nonLeapDays = 0
    if _sDateLeap.month > 2:
        sYear = _sDateLeap.year + 1
    else:
        sYear = _sDateLeap.year
    if _eDateLeap.month < 3:
        eYear = _eDateLeap.year - 1
    else:
        eYear = _eDateLeap.year
    for t in range(sYear, eYear + 1):
        if not cal.isleap(t):
            nonLeapDays += 1
    return nonLeapDays


def padMissingData(x, _sDate, _eDate):
    eDateChkSite = dt.strptime(x["endDate"], "%Y-%m-%d %H:%M:%S").date()
    eDateChkBasin = dt.strptime(_eDate, "%Y-%m-%d").date()

    # sites data ends before the basin's ends
    if eDateChkBasin > eDateChkSite:
        real_days_btw = (eDateChkBasin - eDateChkSite).days
        non_leap_days = nonLeapDaysBetween(eDateChkSite, eDateChkBasin)
        eDiff = (real_days_btw + non_leap_days) + 1
        x["values"] = list(x["values"] + [np.nan] * eDiff)

    sDateChkSite = dt.strptime(x["beginDate"], "%Y-%m-%d %H:%M:%S").date()
    sDateChkBasin = dt.strptime(_sDate, "%Y-%m-%d").date()

    # basins data starts before the sites data
    if sDateChkBasin < sDateChkSite:
        real_days_btw = (sDateChkSite - sDateChkBasin).days
        non_leap_days = nonLeapDaysBetween(sDateChkBasin, sDateChkSite)
        sDiff = real_days_btw + non_leap_days
        x["values"] = list([np.nan] * sDiff + x["values"])

    # basin data starts before site existed or provided data
    if sDateChkBasin > sDateChkSite:
        real_days_btw = (sDateChkBasin - sDateChkSite).days
        non_leap_days = nonLeapDaysBetween(sDateChkSite, sDateChkBasin)
        sDiff = real_days_btw + non_leap_days
        x["values"] = list(x["values"][sDiff:])

    return x


def createPRECProjTrace(i, jDay, lastValue, nanList):
    dailyData = list(i)
    if jDay < 151 and np.isnan(dailyData[151]):
        dailyData[151] = dailyData[150]
    with warnings.catch_warnings():
        warnings.simplefilter("ignore", category=RuntimeWarning)
        if dailyData:
            projection = list(np.nancumsum(np.diff(dailyData[jDay:])))
            projTrace = nanList + [lastValue] + [lastValue + x for x in projection]
        else:
            projTrace = [np.nan] * 366
    return projTrace


def createSWEProjTrace(i, jDay, lastValue, nanList):
    dailyData = list(i)
    if jDay < 151 and np.isnan(dailyData[151]):
        dailyData[151] = dailyData[150]
    with warnings.catch_warnings():
        warnings.simplefilter("ignore", category=RuntimeWarning)
        peakSWEday = np.array(dailyData).argmax()
        meltOut = 0
        for index, day in enumerate(dailyData[peakSWEday:]):
            if day < 0.1:
                meltOut = index + peakSWEday
                break
        if meltOut > jDay:
            projection = list(np.nancumsum(np.diff(dailyData[jDay:meltOut])))
            projTrace = (
                nanList
                + [lastValue]
                + [
                    lastValue + x if abs(x) < lastValue or jDay < peakSWEday else np.nan
                    for x in projection
                ]
            )
            projTrace[:] = [x if x >= 0 else np.nan for x in projTrace]
            meltList = [projTrace[-1]]
            for t in range(0, 366 - len(projTrace) - 1):
                meltRate = (
                    0.00008 * (len(projTrace) + t) ** 2
                    - 0.0159 * (len(projTrace) + t)
                    + 0.7903
                )
                if meltRate < meltList[-1]:
                    meltList.extend([meltList[-1] - meltRate])
                else:
                    meltList.extend([0])
            projTrace.extend(meltList)
        else:
            projTrace = [np.nan] * 366
    return projTrace


def integrateSMS(dataSMS):
    smsAllAvg = {}
    allSites = list(set(list(dataSMS[-8].keys()) + list(dataSMS[-20].keys())))
    for smsSite in allSites:
        sms20 = []
        sms8 = []
        if dataSMS.get(-20):
            sms20 = dataSMS[-20].get(smsSite, [])
        if dataSMS.get(-8):
            sms8 = dataSMS[-8].get(smsSite, [])
        if sms20 and sms8 and len(sms8) == len(sms20):
            smsAll = [sms8, sms20]
            with warnings.catch_warnings():
                warnings.simplefilter("ignore", category=RuntimeWarning)
                smsAvg = np.nanmean(smsAll, axis=0)
        elif sms8:
            sms8[:] = [np.asarray(x) if not math.isnan(x) else np.nan for x in sms8]
            smsAvg = sms8
        elif sms20:
            sms20[:] = [np.asarray(x) if not math.isnan(x) else np.nan for x in sms20]
            smsAvg = sms20
        else:
            smsAvg = []
        smsAllAvg.update({str(smsSite): list(smsAvg)})
    return smsAllAvg


def integrateSTO(dataSTO):
    stoAllAvg = {}
    allSites = list(dataSTO[-2].keys())
    for stoSite in allSites:
        sto8 = []
        sto4 = []
        sto2 = []
        if dataSTO.get(-8):
            sto8 = list(dataSTO[-8].get(stoSite, []))
        if dataSTO.get(-4):
            sto4 = list(dataSTO[-4].get(stoSite, []))
        if dataSTO.get(-2):
            sto2 = list(dataSTO[-2].get(stoSite, []))
        if sto2 and sto4 and sto8:
            sto8[:] = [
                np.asarray(x) * (1 / 3) if not math.isnan(x) else np.nan for x in sto8
            ]
            sto4[:] = [
                np.asarray(x) * (1 / 3) if not math.isnan(x) else np.nan for x in sto4
            ]
            sto2[:] = [
                np.asarray(x) * (1 / 3) if not math.isnan(x) else np.nan for x in sto2
            ]
            stoAll = list([sto8, sto4, sto2])
            stoAvg = [sum(x) for x in zip(*stoAll)]
        elif sto2 and sto8:
            sto8[:] = [
                np.asarray(x) * (1 / 2) if not math.isnan(x) else np.nan for x in sto8
            ]
            sto2[:] = [
                np.asarray(x) * (1 / 2) if not math.isnan(x) else np.nan for x in sto2
            ]
            stoAll = list([sto8, sto2])
            stoAvg = [sum(x) for x in zip(*stoAll)]
        elif sto2 and sto4:
            sto4[:] = [
                np.asarray(x) * (1 / 2) if not math.isnan(x) else np.nan for x in sto4
            ]
            sto2[:] = [
                np.asarray(x) * (1 / 2) if not math.isnan(x) else np.nan for x in sto2
            ]
            stoAll = list([sto4, sto2])
            stoAvg = [sum(x) for x in zip(*stoAll)]
        elif sto4 and sto8:
            sto8[:] = [
                np.asarray(x) * (1 / 2) if not math.isnan(x) else np.nan for x in sto8
            ]
            sto4[:] = [
                np.asarray(x) * (1 / 2) if not math.isnan(x) else np.nan for x in sto4
            ]
            stoAll = list([sto8, sto4])
            stoAvg = [sum(x) for x in zip(*stoAll)]
        elif sto2:
            sto2[:] = [np.asarray(x) if not math.isnan(x) else np.nan for x in sto2]
            stoAvg = list(sto2)
        elif sto4:
            sto4[:] = [np.asarray(x) if not math.isnan(x) else np.nan for x in sto4]
            stoAvg = list(sto4)
        elif sto8:
            sto8[:] = [np.asarray(x) if not math.isnan(x) else np.nan for x in sto8]
            stoAvg = list(sto8)
        else:
            stoAvg = []
        stoAllAvg.update({str(stoSite): list(stoAvg)})
    return stoAllAvg


def trimToOct1(data):
    sDate = dt.strptime(data["beginDate"], "%Y-%m-%d %H:%M:%S")
    if sDate.month == 10 and sDate.day == 1:
        return data
    WY = sDate.year
    if sDate.month > 9:
        WY += 1
    oct1 = dt(WY, 10, 1)
    diff = (oct1 - sDate).days + nonLeapDaysBetween(sDate, oct1)
    data["flags"] = data["flags"][diff:]
    data["values"] = data["values"][diff:]
    data["beginDate"] = oct1.strftime("%Y-%m-%d %H:%M:%S")
    return data


def get_basin_defs(static_dir=STATIC_DIR):
    glob_dir = os.path.join(static_dir, r"basinDefs")
    glob_re = os.path.join(glob_dir, "*.csv")
    basin_defs = []
    for filename in glob.glob(glob_re):
        basin_def_ext = filename.replace(os.path.join(glob_dir, "basinDef_"), "")
        basin_def_ext = basin_def_ext.replace(".csv", "")
        basin_defs.append(basin_def_ext)
    return basin_defs


def getBasinSites(basinName, basinTable):
    siteListStr = basinTable.get(basinName).get(r"BasinSites")
    siteList = []
    if siteListStr:
        siteList = siteListStr.split(r";")
    if siteList:
        siteList[:] = [i.strip() for i in siteList]
        return siteList
    return []


def get_basin_table(state, root_dir):
    delimiter = ","
    basinTable = {}
    basinDefDir = os.path.join(root_dir, r"static", r"basinDefs")
    basinDefFileName = os.path.join(basinDefDir, "basinDef_" + state + r".csv")
    with open(basinDefFileName, "r") as data_file:
        data = csv.reader(data_file, delimiter=delimiter)
        headers = next(data)[1:]
        for row in data:
            temp_dict = {}
            name = row[0]
            values = []
            for x in row[1:]:
                values.append(x)
            for i in range(len(values)):
                temp_dict[headers[i]] = values[i]
            basinTable[name] = temp_dict
    return basinTable


def get_wy_date_series():
    date_series = [
        date(2015, 10, 1) + datetime.timedelta(days=x) for x in range(0, 366)
    ]  # could use any year with a leap day
    return date_series


def get_plot_layout(
    sensor, date_series=None, data_links="", sites_link="", annoData=""
):
    if not date_series:
        date_series = get_wy_date_series()
    layout = go.Layout(
        template="plotly_white",
        images=get_logo(),
        annotations=get_annotations(
            data_links=data_links, sites_link=sites_link, anno_data=annoData
        ),
        legend=dict(
            traceorder="reversed", tracegroupgap=1, bordercolor="#E2E2E2", borderwidth=2
        ),
        showlegend=True,
        autosize=True,
        yaxis=dict(
            title=f"{SENSOR_LONG[sensor]} ({SENSOR_UNITS[sensor]})",
            hoverformat=".1f",
            tickformat=".0f",
            automargin=True,
        ),
        xaxis=dict(
            tickformat="%b %e",
            rangeselector=get_rangeselector(),
            rangeslider=dict(
                thickness=0.05,
                range=[date_series[0], date_series[-1]],
            ),
            type="date",
            fixedrange=True,
            automargin=True,
            range=[date_series[0], date_series[-1]],
        ),
        margin=get_margin(with_title=False),
        dragmode=False,
    )
    return layout


def get_no_chart_fig_data(sensor):
    anno_text = "no_chart"
    layout = get_plot_layout(
        sensor,
        date_series=None,
        data_links="",
        sites_link="",
        annoData="",
    )
    trace = [
        go.Scatter(
            x=get_wy_date_series(),
            y=[np.nan] * len(get_wy_date_series()),
            name="N/A",
            visible=True,
            hoverinfo="none",
            line=dict(color="black"),
        )
    ]
    fig_data = {"data": trace, "layout": layout}
    return fig_data, anno_text


def create_log(path="prodScripts.log"):
    logger = logging.getLogger("awdbPlots rotating log")
    logger.setLevel(logging.INFO)

    handler = TimedRotatingFileHandler(path, when="w6", interval=1, backupCount=4)

    logger.addHandler(handler)


def parse_div_id(html_filename):
    norm_path = os.path.normpath(html_filename.replace(".html", ""))
    img_path = list(reversed(norm_path.split(os.path.sep)))[:4]
    img_path = "-".join(img_path).lower()
    return img_path.replace("assochuc", "").replace(" ", "_")


def get_plot_config(div_id="nrcs_plot"):
    return {
        "modeBarButtonsToRemove": ["sendDataToCloud", "lasso2d", "select2d"],
        "showAxisDragHandles": True,
        "showAxisRangeEntryBoxes": True,
        "displaylogo": False,
        "toImageButtonOptions": {
            "filename": div_id,
            "width": 1200,
            "height": 700,
        },
        # 'scrollZoom': False,
        "modeBarButtonsToAdd": [
            "drawline",
            "drawopenpath",
            "drawclosedpath",
            "drawcircle",
            "drawrect",
            "eraseshape",
            "drawtext",
            "hovercompare",
            "hoverclosest",
        ],
    }


def export_fig_data(fig_data, export_path):
    try:
        date_data = [x.strftime("%m-%d") for x in fig_data["data"][0].x]
        data_dict = {}
        data_dict["date"] = date_data
        y_data = {data.name: list(data.y) for data in fig_data["data"]}
        for k, x in y_data.items():
            if len(x) != 366:
                if k.lower() == "median peak swe":
                    y_data[k] = x + [x[-1]] * (366 - len(x))
                else:
                    y_data[k] = x + [np.nan] * (366 - len(x))
        data_dict = {**data_dict, **y_data}
        df = pd.DataFrame().from_dict(data_dict)
        col_mapper = {}
        # num_sites_row = []
        cols = df.columns
        if any("sites)" in i for i in cols):
            for col in cols:
                if "sites)" in col:
                    col_arr = col.split(" ")
                    col_mapper[col] = col_arr[0]
                    # num_sites_row.append(
                    #     col_arr[1].replace('(', '').replace('sites)', '')
                    # )
                else:
                    col_mapper[col] = col
                    # num_sites_row.append('')
        col_mapper["Stats. Shading"] = "70%"
        df = df.round(decimals=10)
        df.rename(columns=col_mapper, inplace=True)
        df.to_csv(f'{export_path.replace(".html", "")}.csv', index=False)
        df.to_json(
            f'{export_path.replace(".html", "")}.json', orient="records"
        )
    except Exception as e:
        print(f"Could not create data files for {export_path} - {e}")


def export_heatmap_data(fig_data, export_path):
    try:
        site_data = fig_data["data"]
        sms_data = site_data[0]
        date_data = [x.strftime("%m-%d-%Y") for x in sms_data.x]
        data_dict = {}
        data_dict["date"] = date_data
        z_data = {sms_data.y[idx]: data for idx, data in enumerate(sms_data.z)}
        del z_data[0]
        data_dict = {**data_dict, **z_data}
        df = pd.DataFrame().from_dict(data_dict)
        df.to_csv(f'{export_path.replace(".html", "")}.csv', index=False)
        df.to_json(
            f'{export_path.replace(".html", "")}.json', orient="split"
        )

    except Exception as e:
        print(f"Could not create data files for {export_path} - {e}")


def parse_basin_name(basinName):
    if basinName[:2].isdigit():
        return " ".join(basinName.split("_")[1:]).upper()
    return " ".join(basinName.split("_")).upper()


def get_rangeselector():
    return {
        "x": 0.02,
        "xanchor": "left",
        "y": 1.01,
        "yanchor": "bottom",
        "buttons": [{"label": "Reset Range", "step": "all"}],
    }


def get_margin(with_title=False):
    if with_title:
        return {"b": 5, "t": 100, "l": 35, "r": 35, "pad": 4}
    return {"b": 5, "t": 35, "l": 35, "r": 55, "pad": 4}


def get_annotations(data_links="", sites_link="", anno_data=""):
    annos = []
    if data_links:
        annos.append(
            {
                "font": {"size": 10},
                "text": data_links,
                "x": 0.5,
                "y": 1,
                "yref": "paper",
                "xref": "paper",
                "xanchor": "center",
                "yanchor": "bottom",
                "align": "center",
                "showarrow": False,
            }
        )
    if sites_link:
        annos.append(
            {
                "font": {"size": 10},
                "text": sites_link,
                "x": 1.02,
                "y": 1,
                "yref": "paper",
                "xref": "paper",
                "align": "left",
                "xanchor": "left",
                "yanchor": "bottom",
                "showarrow": False,
            }
        )
    if anno_data:
        annos.append(
            {
                "font": {"size": 10},
                "text": anno_data,
                "x": 0.015,
                "y": 0.95,
                "bordercolor": "#A9A9A9",
                "borderwidth": 1,
                "borderpad": 2,
                "yref": "paper",
                "xref": "paper",
                "align": "left",
                "xanchor": "left",
                "yanchor": "top",
                "showarrow": False,
            }
        )

    return annos


def get_logo(nrcs_logo_src=NRCS_LOGO_SRC):
    return [
        {
            "source": nrcs_logo_src,
            "xref": "paper",
            "yref": "paper",
            "x": 1.035,
            "y": -0.025,
            "xanchor": "left",
            "yanchor": "top",
            "sizex": 0.1495,
            "sizey": 0.0975,
            "sizing": "contain",
            "opacity": 0.75,
            "layer": "below",
        }
    ]


def get_data_links(name):
    name = urllib.parse.quote(name)
    return (
        f'Link to data: <a href="./{name}.csv">CSV</a>'
        f"&nbsp;/&nbsp;"
        f'<a href="./{name}.json">JSON</a>'
    )


def get_trace_color(year):
    color_dict = {
        0: "#636EFA",
        1: "#EF553B",
        2: "#00CC96",
        3: "#AB63FA",
        4: "#FFA15A",
        5: "#19D3F3",
        6: "#FF6692",
        7: "#B6E880",
        8: "#FF97FF",
        9: "#FECB52",
    }
    return color_dict.get(int(year) % 10)


def save_prod_plot(fig, plot_name, anno_text, chart_title, plotly_js=PLOTLY_SRC):
    div_id = parse_div_id(plot_name)
    chart_html = plotly_to_html(
        fig=fig,
        config=get_plot_config(div_id=div_id),
        include_plotlyjs=plotly_js,
        div_id=f"plotly-graph-div__{div_id}",
        validate=False,
        full_html=False,
    )
    anno_string = get_anno_str(anno_append=anno_text)
    chart_html = get_template(
        chart_html=chart_html,
        chart_title=chart_title,
        anno_string=anno_string,
        css=get_css(),
        div_id=div_id,
    )
    with open(plot_name, "w") as chart:
        chart.write(chart_html)


def save_dev_plot(fig, plot_name, anno_text, chart_title):
    div_id = parse_div_id(plot_name)
    fig.update_layout(title=chart_title, margin=get_margin(with_title=True))
    py.plot(
        fig,
        filename=plot_name,
        auto_open=False,
        show_link=False,
        include_plotlyjs="cdn",
        config=get_plot_config(div_id=div_id),
        validate=False,
    )
    add_cdn(plot_name, anno_append=anno_text)


def parse_args():
    cli_desc = """
    Creates awdb based charts
    """
    parser = argparse.ArgumentParser(description=cli_desc)
    parser.add_argument(
        "-V", "--version", action="store_true", help="show program version"
    )
    parser.add_argument(
        "-o",
        "--output",
        default="local",
        type=str,
        help="sets the base output directory, defaults to local project base",
    )
    parser.add_argument(
        "-u",
        "--url",
        default=DEFAULT_DOMAIN,
        type=str,
        help="sets the URL path to the data files directory",
    )
    parser.add_argument(
        "-p",
        "--path",
        default=None,
        type=str,
        help="sets the file path to the data files directory",
    )
    parser.add_argument(
        "-d",
        "--data_source",
        choices=["url", "file"],
        default="url",
        type=str,
        help="determines whether the script accesses data through URL or file system",
    )
    parser.add_argument(
        "--dev",
        action="store_true",
        help="Create chart without external dependencies",
    )
    args = parser.parse_args()
    return args


if __name__ == "__main__":
    a = get_basin_defs()
    print("why are you running this?")
    dirpath = os.getcwd()
    print(os.path.basename(dirpath))



import datetime
import math
import warnings
from os import makedirs, path

import numpy as np
import plotly.graph_objs as go
from utils.awdb import POR_ONLY_TXT  # universal dicts
from utils.awdb import create_log  # setup for logger
from utils.awdb import get_site_data  # url GET functions
from utils.awdb import (  # universal datetime items; only needed for basin charts; unique functions
    BASIN_NETWORKS,
    SENSOR_LONG,
    date,
    dt,
    export_fig_data,
    get_basin_defs,
    get_basin_sites_link,
    get_basin_table,
    get_data_links,
    get_plot_layout,
    get_trace_color,
    get_wy_date_series,
    getBasinSites,
    is_leap_wy,
    padMissingData,
    parse_args,
    parse_basin_name,
    save_dev_plot,
    save_prod_plot,
    today,
    trimToOct1,
    get_no_chart_fig_data,
)


def update_chart(basin, basinSites, all_meta, args):
    print("Working on WTEQ POR Chart for " + basinName)
    statsData = []
    minData = []
    maxData = []
    meanData = []
    lowestData = []
    highestData = []
    lowData = []
    highData = []
    meanData = []
    trace = []
    plotData = []
    basinPlotData = []
    PORplotData = []
    basinNormData = []
    basinPlotNormData = []
    validTrip = []

    networks = BASIN_NETWORKS
    sensor = args.sensor

    meta = [
        x
        for x in all_meta
        if str.split(x["stationTriplet"], ":")[2] in networks
        and str.split(x["stationTriplet"], ":")[0] in basinSites
    ]

    validTrip = [x["stationTriplet"] for x in meta]
    date_series = get_wy_date_series()

    if validTrip:
        normData = []
        urls = []
        for triplet in validTrip:
            url = "/".join(
                [
                    "sitedata",
                    "normals",
                    "DAILY",
                    sensor,
                    triplet.replace(":", "_") + ".json",
                ]
            )
            urls.append(url)
        normData = get_site_data(urls, validTrip, args)

        basinNormData = [
            np.array(x["values"], dtype=float)
            for x in normData
            if len(x["values"]) == 366 and not np.nan in x["values"]
        ]
        sites_with_norms = [x["id"] for x in normData if x["values"]]
    else:
        print(f"     No sites with {sensor} data in {basin}, skipping...")
        return None, ""

    if basinNormData:
        with warnings.catch_warnings():
            warnings.simplefilter("ignore", category=RuntimeWarning)
            basinPlotNormData = list(
                np.nanmean(np.array([i for i in basinNormData]), axis=0)
            )
        validTrip[:] = [x for x in validTrip if x in sites_with_norms]

    if not validTrip:
        print(f"     No sites with {sensor} norm data in {basin}, skipping...")
        return None, ""

    data = []
    urls = []
    for triplet in validTrip:
        url = "/".join(
            ["sitedata", "DAILY", sensor, triplet.replace(":", "_") + ".json"]
        )
        urls.append(url)

    data = get_site_data(urls, validTrip, args)
    data[:] = [trimToOct1(i) for i in data if i]

    beginDateDict = {}
    for siteMeta in data:
        beginDateDict.update(
            {
                str(siteMeta["stationTriplet"]): dt.strptime(
                    str(siteMeta["beginDate"]), "%Y-%m-%d %H:%M:%S"
                )
            }
        )
    if len(beginDateDict.values()) == 0:
        print(f"     No sites with {sensor} data in {basin}, skipping...")
        return None, ""
    basinBeginDate = min(beginDateDict.values())

    sYear = basinBeginDate.year
    if basinBeginDate.year > sYear:
        if basinBeginDate.month < 10:
            sYear = basinBeginDate.year
        else:
            if basinBeginDate.month == 10 and basinBeginDate.day == 1:
                sYear = basinBeginDate.year
            else:
                sYear = basinBeginDate.year + 1

    sDate = date(sYear, 10, 1).strftime("%Y-%m-%d")
    eDate = today.date().strftime("%Y-%m-%d")

    for dataSite in data:
        if dataSite:
            padMissingData(dataSite, sDate, eDate)

    plotData = [np.array(x["values"], dtype=float) for x in data if x]
    plotData[:] = [x for x in plotData if not np.isnan(x).all()]

    if not plotData:
        print(f"     No data for {sensor} in {basin}, skipping...")
        return None, ""

    with warnings.catch_warnings():
        warnings.simplefilter("ignore", category=RuntimeWarning)
        basinPlotData = list(np.nanmean(np.array(plotData), axis=0))
        nonNan = [[0 if np.isnan(x) else 1 for x in i] for i in plotData]
        nonNanSum = [np.nansum(i) for i in zip(*nonNan)]
        dailySitesNum = list(
            [nonNanSum[i : i + 366] for i in range(0, len(nonNanSum), 366)]
        )
        yearlySitesNum = [int(np.nanmedian(i)) for i in dailySitesNum]

    PORplotData = list(
        [basinPlotData[i : i + 366] for i in range(0, len(basinPlotData), 366)]
    )

    if not yearlySitesNum:
        print(f"     No data for {sensor} in {basin}, skipping...")
        return None, ""

    maxSites = max(yearlySitesNum)
    validYears = [
        sYear + i + 1 for i, x in enumerate(yearlySitesNum) if x > maxSites * 0.5
    ]
    PORplotData[:] = [
        PORplotData[i] for i, x in enumerate(yearlySitesNum) if x > maxSites * 0.5
    ]

    percentileData = list(PORplotData)
    allButCurrWY = list(PORplotData)
    validYearlySitesNum = [x for x in yearlySitesNum if x > maxSites * 0.5]

    if not allButCurrWY:
        print(f"     No data for {sensor} in {basin}, skipping...")
        return None, ""

    del allButCurrWY[-1]
    with warnings.catch_warnings():
        warnings.simplefilter("ignore", category=RuntimeWarning)
        maxSWE = [np.nanmax(x) for x in allButCurrWY]
        jDayMaxSWE = [
            np.argwhere(x == maxSWE[i])[-1]
            for i, x in enumerate(allButCurrWY)
            if np.argwhere(x == maxSWE[i]).any()
        ]
        medianjDayMaxSWE = np.nanpercentile(jDayMaxSWE, 50)
        if is_leap_wy(today) and medianjDayMaxSWE > 151:
            medianjDayMaxSWE += -1
        if np.isnan(medianjDayMaxSWE):
            medianjDayMaxSWE = 0
        medianDateMaxSWE = date_series[0] + datetime.timedelta(days=medianjDayMaxSWE)
        medianMaxSWE = np.nanpercentile(maxSWE, 50)
    percentileData[-1] = np.append(
        percentileData[-1], [np.nan] * (366 - len(percentileData[-1]))
    )
    percentileData[:] = list(map(list, zip(*percentileData)))
    statsData = list(map(list, zip(*allButCurrWY)))

    if len(percentileData[0]) > 1:
        percentileData[151] = percentileData[150]

    if len(statsData) > 1:
        statsData[151] = statsData[150]
        with warnings.catch_warnings():
            warnings.simplefilter("ignore", category=RuntimeWarning)
            minData = [np.nanmin(a) for a in statsData]
            maxData = [np.nanmax(a) for a in statsData]
            meanData = [np.nanpercentile(a, 50) for a in statsData]
            lowestData = [np.nanpercentile(a, 10) for a in statsData]
            highestData = [np.nanpercentile(a, 90) for a in statsData]
            lowData = [np.nanpercentile(a, 30) for a in statsData]
            highData = [np.nanpercentile(a, 70) for a in statsData]

    show_trace = {True: "legendonly", False: False}
    used_in_stats = {True: "", False: "<sup>+</sup>"}

    if len(PORplotData) > 0:
        for index, i in enumerate(PORplotData):
            name = (
                f"{validYears[index]} "
                f"({validYearlySitesNum[index]} sites)"
                f"{used_in_stats[validYearlySitesNum[index] > maxSites*0.5]}"
            )
            if index == len(PORplotData) - 1:
                trace.extend(
                    [
                        go.Scatter(
                            x=date_series,
                            y=i,
                            name=f"{name}",
                            visible=True,
                            connectgaps=False,
                            line=dict(color="rgb(0,0,0)"),
                            mode="lines+markers",
                            marker_size=2,
                        )
                    ]
                )
            elif np.nansum(i) > 0:
                trace.extend(
                    [
                        go.Scatter(
                            x=date_series,
                            y=i,
                            name=f"{name}",
                            visible=show_trace[
                                int(validYearlySitesNum[index]) > maxSites * 0.5
                            ],
                            connectgaps=False,
                            mode="lines+markers",
                            marker_size=2,
                            line=dict(color=get_trace_color(index)),
                        )
                    ]
                )
    if meanData:
        if lowestData:
            trace.extend(
                [
                    go.Scatter(
                        x=date_series,
                        y=minData,
                        legendgroup="centiles",
                        name=r"Min",
                        visible=True,
                        mode="lines",
                        line=dict(width=0),
                        connectgaps=True,
                        fillcolor="rgba(237,0,1,0.15)",
                        fill="none",
                        showlegend=False,
                        hoverinfo="none",
                    )
                ]
            )
            trace.extend(
                [
                    go.Scatter(
                        x=date_series,
                        y=lowestData,
                        legendgroup="centiles",
                        name=r"10%",
                        visible=True,
                        mode="lines",
                        line=dict(width=0),
                        connectgaps=True,
                        fillcolor="rgba(237,0,1,0.15)",
                        fill="tonexty",
                        showlegend=False,
                        hoverinfo="none",
                    )
                ]
            )
        if lowData:
            trace.extend(
                [
                    go.Scatter(
                        x=date_series,
                        y=lowData,
                        legendgroup="centiles",
                        name=r"30%",
                        visible=True,
                        mode="lines",
                        line=dict(width=0),
                        connectgaps=True,
                        fillcolor="rgba(237,237,0,0.15)",
                        fill="tonexty",
                        showlegend=False,
                        hoverinfo="none",
                    )
                ]
            )
        if highData:
            trace.extend(
                [
                    go.Scatter(
                        x=date_series,
                        y=highData,
                        legendgroup="centiles",
                        name=r"Stats. Shading",
                        visible=True,
                        mode="lines",
                        line=dict(width=0),
                        connectgaps=True,
                        fillcolor="rgba(115,237,115,0.15)",
                        fill="tonexty",
                        showlegend=True,
                        hoverinfo="none",
                    )
                ]
            )
        if highestData:
            trace.extend(
                [
                    go.Scatter(
                        x=date_series,
                        y=highestData,
                        legendgroup="centiles",
                        connectgaps=True,
                        name=r"90%",
                        visible=True,
                        mode="lines",
                        line=dict(width=0),
                        fillcolor="rgba(0,237,237,0.15)",
                        fill="tonexty",
                        showlegend=False,
                        hoverinfo="none",
                    )
                ]
            )
            trace.extend(
                [
                    go.Scatter(
                        x=date_series,
                        y=maxData,
                        legendgroup="centiles",
                        name=r"Max",
                        visible=True,
                        mode="lines",
                        line=dict(width=0),
                        connectgaps=True,
                        fillcolor="rgba(1,0,237,0.15)",
                        fill="tonexty",
                        showlegend=False,
                        hoverinfo="none",
                    )
                ]
            )
    if minData:
        trace.extend(
            [
                go.Scatter(
                    x=date_series,
                    y=minData,
                    name=r"Min",
                    visible=True,
                    # hoverinfo='none',
                    connectgaps=True,
                    line=dict(color="rgba(237,0,0,0.5)"),
                )
            ]
        )

    if basinPlotNormData:
        trace.extend(
            [
                go.Scatter(
                    x=date_series,
                    y=basinPlotNormData,
                    name=r"Median ('91-'20)",
                    connectgaps=True,
                    visible=True,
                    # hoverinfo='none',
                    line=dict(color="rgba(0,237,0,0.4)"),
                )
            ]
        )

    if meanData:
        if basinPlotNormData:
            trace.extend(
                [
                    go.Scatter(
                        x=date_series,
                        y=meanData,
                        name=r"Median (POR)",
                        visible="legendonly",
                        # hoverinfo='none',
                        connectgaps=True,
                        line=dict(color="rgba(0,237,0,0.4)", dash="dash"),
                    )
                ]
            )
        else:
            trace.extend(
                [
                    go.Scatter(
                        x=date_series,
                        y=meanData,
                        name=r"Median (POR)",
                        connectgaps=True,
                        visible=True,
                        # hoverinfo='none',
                        line=dict(color="rgba(0,237,0,0.4)"),
                    )
                ]
            )
    if maxData:
        trace.extend(
            [
                go.Scatter(
                    x=date_series,
                    y=maxData,
                    name=r"Max",
                    visible=True,
                    # hoverinfo='none',
                    connectgaps=True,
                    line=dict(color="rgba(0,0,237,0.4)"),
                )
            ]
        )

    if medianMaxSWE:
        trace.extend(
            [
                go.Scatter(
                    x=[medianDateMaxSWE],
                    y=[medianMaxSWE],
                    name="Median Peak SWE",
                    visible=True,
                    mode="markers",
                    marker=dict(size=8, color="rgba(0,237,0,0.75)", symbol="x"),
                )
            ]
        )

    sites_link = get_basin_sites_link(validTrip)
    asterisk = ""
    anno_text = ""
    if not basinPlotNormData:
        basinPlotNormData = meanData
        if len(basinPlotNormData) == 0:
            basinPlotNormData = [0] * 366
        anno_text = f"{anno_text}{POR_ONLY_TXT}"
        asterisk = "*"
    jDay = len(PORplotData[-1]) - 1

    if basinPlotNormData[jDay] == 0:
        perNorm = r"N/A"
        perPeak = r"N/A"
    elif np.isnan(PORplotData[-1][jDay]):
        perNorm = r"N/A"
        perPeak = r"N/A"
    else:
        perNorm = str(
            "{0:g}".format(
                100 * round(PORplotData[-1][jDay] / basinPlotNormData[jDay], 2)
            )
        )
        perPeak = str(
            "{0:g}".format(100 * round(PORplotData[-1][jDay] / medianMaxSWE, 2))
        )

    if not math.isnan(PORplotData[-1][jDay]) and statsData:
        jDayData = [x for x in percentileData[jDay] if not np.isnan(x)]
        jDayData.sort(reverse=True)
        rank = jDayData.index(PORplotData[-1][jDay]) + 1
        centile = str(int(round(100 * (1 - ((rank - 1) / (len(jDayData) - 1))), 0)))
    else:
        centile = "N/A"

    dayOfPeak = int(medianjDayMaxSWE)
    if jDay > dayOfPeak:
        tense = r"Since"
    else:
        tense = r"Until"
    daysToPeak = str(abs(jDay - dayOfPeak))
    anno_data = str(
        r"Current"
        + asterisk
        + " as of "
        + today.strftime("%m/%d/%Y")
        + ":<br>% of Median - "
        + perNorm
        + r"%<br>"
        + r"% Median Peak - "
        + perPeak
        + r"%<br>"
        + r"Days "
        + tense
        + r" Median Peak - "
        + daysToPeak
        + r"<br>"
        r"Percentile - " + centile
    )

    data_links = get_data_links(basinName)

    layout = get_plot_layout(
        sensor=sensor, 
        date_series=date_series, 
        data_links=data_links, 
        sites_link=sites_link, 
        annoData=anno_data
    )

    return {"data": trace, "layout": layout}, anno_text


if __name__ == "__main__":
    import logging
    import sys
    import time
    import traceback

    args = parse_args()

    if args.version:
        print("v1.0")

    scriptName = path.basename(__file__)
    this_dir = path.dirname(path.abspath(__file__))
    root_dir = path.dirname(this_dir)
    if args.output == "local":
        master_dir = path.dirname(this_dir)
    elif path.isdir(args.output):
        master_dir = args.output
    else:
        print(f"Invalid output directory ({args.output}), please try again...")
        sys.exit(1)

    create_log()
    script_name_str = scriptName.replace("_", " ").replace(".py", "")
    charts_count = 0
    err_count = 0

    states = get_basin_defs()
    sensor = "WTEQ"
    args.sensor = sensor
    url = "/".join(["sitedata", "metadata", sensor, "metadata.json"])
    all_meta = get_site_data([url], [f"{sensor}-meta"], args)[0]
    for state in states:
        relPath = ["basin-plots", "POR", sensor, state]
        dirPath = path.join(master_dir, *relPath)
        makedirs(dirPath, exist_ok=True)
        basinTable = get_basin_table(state, root_dir)
        for basinName in basinTable:
            bt = time.time()
            plot_name = path.join(dirPath, basinName + r".html")
            basin = parse_basin_name(basinName).upper()
            chart_title = f"{SENSOR_LONG[sensor]} in {basin}".upper()
            try:
                basinSites = getBasinSites(basinName, basinTable)
                fig_data, anno_text = update_chart(basinName, basinSites, all_meta, args)
                if not fig_data:
                    fig_data, anno_text = get_no_chart_fig_data(sensor)

                export_fig_data(fig_data, plot_name)
                fig = go.Figure(fig_data)
                if args.dev:
                    save_dev_plot(fig, plot_name, anno_text, chart_title)
                else:
                    save_prod_plot(fig, plot_name, anno_text, chart_title)
                charts_count += 1
                chart_pub_str = (
                    f"     {basinName} {script_name_str} "
                    f"Chart created in "
                    f"{round(time.time()-bt,2)} seconds"
                )
                print(chart_pub_str)

            except Exception as e:
                exc_type, exc_value, exc_traceback = sys.exc_info()
                error_trace = traceback.format_tb(exc_traceback)
                error_line = error_trace[-1].split(",")[1].strip()
                e = str(e).replace(",", " ")
                err_str = (
                    f"     No {script_name_str} "
                    f"chart created for {basinName} "
                    f"- {e} - {error_line}"
                )
                print(err_str)

                logging_str = (
                    f"CHART FAILED,"
                    f'{scriptName.replace("_",",")},'
                    f'{state},{basinName.replace(","," ")},{e},'
                    f'{error_line.replace("line ","")},'
                    f'{today.strftime("%Y-%m-%d,%H:%M:%S")}'
                )

                logging.error(logging_str)
                err_count += 1

    logging_str = (
        f"# CHARTS PRODUCED,"
        f"{charts_count},"
        f'{scriptName.replace("_",",")},'
        f'{today.strftime("%Y-%m-%d,%H:%M:%S")}'
    )
    logging.info(logging_str)
    logging_str = (
        f"# CHARTS FAILED,"
        f"{err_count},"
        f'{scriptName.replace("_",",")},'
        f'{today.strftime("%Y-%m-%d,%H:%M:%S")}'
    )
    logging.info(logging_str)
```

```python
import datetime
import math
import warnings
from os import makedirs, path

import numpy as np
import plotly.graph_objs as go
# from utils.awdb import POR_ONLY_TXT  # universal dicts
# from utils.awdb import create_log  # setup for logger
# from utils.awdb import get_site_data  # url GET functions
# from utils.awdb import (  # universal datetime items; only needed for basin charts; unique functions
#     BASIN_NETWORKS,
#     SENSOR_LONG,
#     date,
#     dt,
#     export_fig_data,
#     get_basin_defs,
#     get_basin_sites_link,
#     get_basin_table,
#     get_data_links,
#     get_plot_layout,
#     get_trace_color,
#     get_wy_date_series,
#     getBasinSites,
#     is_leap_wy,
#     padMissingData,
#     parse_args,
#     parse_basin_name,
#     save_dev_plot,
#     save_prod_plot,
#     today,
#     trimToOct1,
#     get_no_chart_fig_data,
# )


def update_chart(basin, basinSites, all_meta, args):
    print("Working on WTEQ POR Chart for " + basinName)
    statsData = []
    minData = []
    maxData = []
    meanData = []
    lowestData = []
    highestData = []
    lowData = []
    highData = []
    meanData = []
    trace = []
    plotData = []
    basinPlotData = []
    PORplotData = []
    basinNormData = []
    basinPlotNormData = []
    validTrip = []

    networks = BASIN_NETWORKS
    sensor = args.sensor

    meta = [
        x
        for x in all_meta
        if str.split(x["stationTriplet"], ":")[2] in networks
        and str.split(x["stationTriplet"], ":")[0] in basinSites
    ]

    validTrip = [x["stationTriplet"] for x in meta]
    date_series = get_wy_date_series()

    if validTrip:
        normData = []
        urls = []
        for triplet in validTrip:
            url = "/".join(
                [
                    "sitedata",
                    "normals",
                    "DAILY",
                    sensor,
                    triplet.replace(":", "_") + ".json",
                ]
            )
            urls.append(url)
        normData = get_site_data(urls, validTrip, args)

        basinNormData = [
            np.array(x["values"], dtype=float)
            for x in normData
            if len(x["values"]) == 366 and not np.nan in x["values"]
        ]
        sites_with_norms = [x["id"] for x in normData if x["values"]]
    else:
        print(f"     No sites with {sensor} data in {basin}, skipping...")
        return None, ""

    if basinNormData:
        with warnings.catch_warnings():
            warnings.simplefilter("ignore", category=RuntimeWarning)
            basinPlotNormData = list(
                np.nanmean(np.array([i for i in basinNormData]), axis=0)
            )
        validTrip[:] = [x for x in validTrip if x in sites_with_norms]

    if not validTrip:
        print(f"     No sites with {sensor} norm data in {basin}, skipping...")
        return None, ""

    data = []
    urls = []
    for triplet in validTrip:
        url = "/".join(
            ["sitedata", "DAILY", sensor, triplet.replace(":", "_") + ".json"]
        )
        urls.append(url)

    data = get_site_data(urls, validTrip, args)
    data[:] = [trimToOct1(i) for i in data if i]

    beginDateDict = {}
    for siteMeta in data:
        beginDateDict.update(
            {
                str(siteMeta["stationTriplet"]): dt.strptime(
                    str(siteMeta["beginDate"]), "%Y-%m-%d %H:%M:%S"
                )
            }
        )
    if len(beginDateDict.values()) == 0:
        print(f"     No sites with {sensor} data in {basin}, skipping...")
        return None, ""
    basinBeginDate = min(beginDateDict.values())

    sYear = basinBeginDate.year
    if basinBeginDate.year > sYear:
        if basinBeginDate.month < 10:
            sYear = basinBeginDate.year
        else:
            if basinBeginDate.month == 10 and basinBeginDate.day == 1:
                sYear = basinBeginDate.year
            else:
                sYear = basinBeginDate.year + 1

    sDate = date(sYear, 10, 1).strftime("%Y-%m-%d")
    eDate = today.date().strftime("%Y-%m-%d")

    for dataSite in data:
        if dataSite:
            padMissingData(dataSite, sDate, eDate)

    plotData = [np.array(x["values"], dtype=float) for x in data if x]
    plotData[:] = [x for x in plotData if not np.isnan(x).all()]

    if not plotData:
        print(f"     No data for {sensor} in {basin}, skipping...")
        return None, ""

    with warnings.catch_warnings():
        warnings.simplefilter("ignore", category=RuntimeWarning)
        basinPlotData = list(np.nanmean(np.array(plotData), axis=0))
        nonNan = [[0 if np.isnan(x) else 1 for x in i] for i in plotData]
        nonNanSum = [np.nansum(i) for i in zip(*nonNan)]
        dailySitesNum = list(
            [nonNanSum[i : i + 366] for i in range(0, len(nonNanSum), 366)]
        )
        yearlySitesNum = [int(np.nanmedian(i)) for i in dailySitesNum]

    PORplotData = list(
        [basinPlotData[i : i + 366] for i in range(0, len(basinPlotData), 366)]
    )

    if not yearlySitesNum:
        print(f"     No data for {sensor} in {basin}, skipping...")
        return None, ""

    maxSites = max(yearlySitesNum)
    validYears = [
        sYear + i + 1 for i, x in enumerate(yearlySitesNum) if x > maxSites * 0.5
    ]
    PORplotData[:] = [
        PORplotData[i] for i, x in enumerate(yearlySitesNum) if x > maxSites * 0.5
    ]

    percentileData = list(PORplotData)
    allButCurrWY = list(PORplotData)
    validYearlySitesNum = [x for x in yearlySitesNum if x > maxSites * 0.5]

    if not allButCurrWY:
        print(f"     No data for {sensor} in {basin}, skipping...")
        return None, ""

    del allButCurrWY[-1]
    with warnings.catch_warnings():
        warnings.simplefilter("ignore", category=RuntimeWarning)
        maxSWE = [np.nanmax(x) for x in allButCurrWY]
        jDayMaxSWE = [
            np.argwhere(x == maxSWE[i])[-1]
            for i, x in enumerate(allButCurrWY)
            if np.argwhere(x == maxSWE[i]).any()
        ]
        medianjDayMaxSWE = np.nanpercentile(jDayMaxSWE, 50)
        if is_leap_wy(today) and medianjDayMaxSWE > 151:
            medianjDayMaxSWE += -1
        if np.isnan(medianjDayMaxSWE):
            medianjDayMaxSWE = 0
        medianDateMaxSWE = date_series[0] + datetime.timedelta(days=medianjDayMaxSWE)
        medianMaxSWE = np.nanpercentile(maxSWE, 50)
    percentileData[-1] = np.append(
        percentileData[-1], [np.nan] * (366 - len(percentileData[-1]))
    )
    percentileData[:] = list(map(list, zip(*percentileData)))
    statsData = list(map(list, zip(*allButCurrWY)))

    if len(percentileData[0]) > 1:
        percentileData[151] = percentileData[150]

    if len(statsData) > 1:
        statsData[151] = statsData[150]
        with warnings.catch_warnings():
            warnings.simplefilter("ignore", category=RuntimeWarning)
            minData = [np.nanmin(a) for a in statsData]
            maxData = [np.nanmax(a) for a in statsData]
            meanData = [np.nanpercentile(a, 50) for a in statsData]
            lowestData = [np.nanpercentile(a, 10) for a in statsData]
            highestData = [np.nanpercentile(a, 90) for a in statsData]
            lowData = [np.nanpercentile(a, 30) for a in statsData]
            highData = [np.nanpercentile(a, 70) for a in statsData]

    show_trace = {True: "legendonly", False: False}
    used_in_stats = {True: "", False: "<sup>+</sup>"}

    if len(PORplotData) > 0:
        for index, i in enumerate(PORplotData):
            name = (
                f"{validYears[index]} "
                f"({validYearlySitesNum[index]} sites)"
                f"{used_in_stats[validYearlySitesNum[index] > maxSites*0.5]}"
            )
            if index == len(PORplotData) - 1:
                trace.extend(
                    [
                        go.Scatter(
                            x=date_series,
                            y=i,
                            name=f"{name}",
                            visible=True,
                            connectgaps=False,
                            line=dict(color="rgb(0,0,0)"),
                            mode="lines+markers",
                            marker_size=2,
                        )
                    ]
                )
            elif np.nansum(i) > 0:
                trace.extend(
                    [
                        go.Scatter(
                            x=date_series,
                            y=i,
                            name=f"{name}",
                            visible=show_trace[
                                int(validYearlySitesNum[index]) > maxSites * 0.5
                            ],
                            connectgaps=False,
                            mode="lines+markers",
                            marker_size=2,
                            line=dict(color=get_trace_color(index)),
                        )
                    ]
                )
    if meanData:
        if lowestData:
            trace.extend(
                [
                    go.Scatter(
                        x=date_series,
                        y=minData,
                        legendgroup="centiles",
                        name=r"Min",
                        visible=True,
                        mode="lines",
                        line=dict(width=0),
                        connectgaps=True,
                        fillcolor="rgba(237,0,1,0.15)",
                        fill="none",
                        showlegend=False,
                        hoverinfo="none",
                    )
                ]
            )
            trace.extend(
                [
                    go.Scatter(
                        x=date_series,
                        y=lowestData,
                        legendgroup="centiles",
                        name=r"10%",
                        visible=True,
                        mode="lines",
                        line=dict(width=0),
                        connectgaps=True,
                        fillcolor="rgba(237,0,1,0.15)",
                        fill="tonexty",
                        showlegend=False,
                        hoverinfo="none",
                    )
                ]
            )
        if lowData:
            trace.extend(
                [
                    go.Scatter(
                        x=date_series,
                        y=lowData,
                        legendgroup="centiles",
                        name=r"30%",
                        visible=True,
                        mode="lines",
                        line=dict(width=0),
                        connectgaps=True,
                        fillcolor="rgba(237,237,0,0.15)",
                        fill="tonexty",
                        showlegend=False,
                        hoverinfo="none",
                    )
                ]
            )
        if highData:
            trace.extend(
                [
                    go.Scatter(
                        x=date_series,
                        y=highData,
                        legendgroup="centiles",
                        name=r"Stats. Shading",
                        visible=True,
                        mode="lines",
                        line=dict(width=0),
                        connectgaps=True,
                        fillcolor="rgba(115,237,115,0.15)",
                        fill="tonexty",
                        showlegend=True,
                        hoverinfo="none",
                    )
                ]
            )
        if highestData:
            trace.extend(
                [
                    go.Scatter(
                        x=date_series,
                        y=highestData,
                        legendgroup="centiles",
                        connectgaps=True,
                        name=r"90%",
                        visible=True,
                        mode="lines",
                        line=dict(width=0),
                        fillcolor="rgba(0,237,237,0.15)",
                        fill="tonexty",
                        showlegend=False,
                        hoverinfo="none",
                    )
                ]
            )
            trace.extend(
                [
                    go.Scatter(
                        x=date_series,
                        y=maxData,
                        legendgroup="centiles",
                        name=r"Max",
                        visible=True,
                        mode="lines",
                        line=dict(width=0),
                        connectgaps=True,
                        fillcolor="rgba(1,0,237,0.15)",
                        fill="tonexty",
                        showlegend=False,
                        hoverinfo="none",
                    )
                ]
            )
    if minData:
        trace.extend(
            [
                go.Scatter(
                    x=date_series,
                    y=minData,
                    name=r"Min",
                    visible=True,
                    # hoverinfo='none',
                    connectgaps=True,
                    line=dict(color="rgba(237,0,0,0.5)"),
                )
            ]
        )

    if basinPlotNormData:
        trace.extend(
            [
                go.Scatter(
                    x=date_series,
                    y=basinPlotNormData,
                    name=r"Median ('91-'20)",
                    connectgaps=True,
                    visible=True,
                    # hoverinfo='none',
                    line=dict(color="rgba(0,237,0,0.4)"),
                )
            ]
        )

    if meanData:
        if basinPlotNormData:
            trace.extend(
                [
                    go.Scatter(
                        x=date_series,
                        y=meanData,
                        name=r"Median (POR)",
                        visible="legendonly",
                        # hoverinfo='none',
                        connectgaps=True,
                        line=dict(color="rgba(0,237,0,0.4)", dash="dash"),
                    )
                ]
            )
        else:
            trace.extend(
                [
                    go.Scatter(
                        x=date_series,
                        y=meanData,
                        name=r"Median (POR)",
                        connectgaps=True,
                        visible=True,
                        # hoverinfo='none',
                        line=dict(color="rgba(0,237,0,0.4)"),
                    )
                ]
            )
    if maxData:
        trace.extend(
            [
                go.Scatter(
                    x=date_series,
                    y=maxData,
                    name=r"Max",
                    visible=True,
                    # hoverinfo='none',
                    connectgaps=True,
                    line=dict(color="rgba(0,0,237,0.4)"),
                )
            ]
        )

    if medianMaxSWE:
        trace.extend(
            [
                go.Scatter(
                    x=[medianDateMaxSWE],
                    y=[medianMaxSWE],
                    name="Median Peak SWE",
                    visible=True,
                    mode="markers",
                    marker=dict(size=8, color="rgba(0,237,0,0.75)", symbol="x"),
                )
            ]
        )

    sites_link = get_basin_sites_link(validTrip)
    asterisk = ""
    anno_text = ""
    if not basinPlotNormData:
        basinPlotNormData = meanData
        if len(basinPlotNormData) == 0:
            basinPlotNormData = [0] * 366
        anno_text = f"{anno_text}{POR_ONLY_TXT}"
        asterisk = "*"
    jDay = len(PORplotData[-1]) - 1

    if basinPlotNormData[jDay] == 0:
        perNorm = r"N/A"
        perPeak = r"N/A"
    elif np.isnan(PORplotData[-1][jDay]):
        perNorm = r"N/A"
        perPeak = r"N/A"
    else:
        perNorm = str(
            "{0:g}".format(
                100 * round(PORplotData[-1][jDay] / basinPlotNormData[jDay], 2)
            )
        )
        perPeak = str(
            "{0:g}".format(100 * round(PORplotData[-1][jDay] / medianMaxSWE, 2))
        )

    if not math.isnan(PORplotData[-1][jDay]) and statsData:
        jDayData = [x for x in percentileData[jDay] if not np.isnan(x)]
        jDayData.sort(reverse=True)
        rank = jDayData.index(PORplotData[-1][jDay]) + 1
        centile = str(int(round(100 * (1 - ((rank - 1) / (len(jDayData) - 1))), 0)))
    else:
        centile = "N/A"

    dayOfPeak = int(medianjDayMaxSWE)
    if jDay > dayOfPeak:
        tense = r"Since"
    else:
        tense = r"Until"
    daysToPeak = str(abs(jDay - dayOfPeak))
    anno_data = str(
        r"Current"
        + asterisk
        + " as of "
        + today.strftime("%m/%d/%Y")
        + ":<br>% of Median - "
        + perNorm
        + r"%<br>"
        + r"% Median Peak - "
        + perPeak
        + r"%<br>"
        + r"Days "
        + tense
        + r" Median Peak - "
        + daysToPeak
        + r"<br>"
        r"Percentile - " + centile
    )

    data_links = get_data_links(basinName)

    layout = get_plot_layout(
        sensor=sensor, 
        date_series=date_series, 
        data_links=data_links, 
        sites_link=sites_link, 
        annoData=anno_data
    )

    return {"data": trace, "layout": layout}, anno_text


if __name__ == "__main__":
    import logging
    import sys
    import time
    import traceback

    args = parse_args()

    if args.version:
        print("v1.0")

    scriptName = path.basename(__file__)
    this_dir = path.dirname(path.abspath(__file__))
    root_dir = path.dirname(this_dir)
    if args.output == "local":
        master_dir = path.dirname(this_dir)
    elif path.isdir(args.output):
        master_dir = args.output
    else:
        print(f"Invalid output directory ({args.output}), please try again...")
        sys.exit(1)

    create_log()
    script_name_str = scriptName.replace("_", " ").replace(".py", "")
    charts_count = 0
    err_count = 0

    states = get_basin_defs()
    sensor = "WTEQ"
    args.sensor = sensor
    url = "/".join(["sitedata", "metadata", sensor, "metadata.json"])
    all_meta = get_site_data([url], [f"{sensor}-meta"], args)[0]
    for state in states:
        relPath = ["basin-plots", "POR", sensor, state]
        dirPath = path.join(master_dir, *relPath)
        makedirs(dirPath, exist_ok=True)
        basinTable = get_basin_table(state, root_dir)
        for basinName in basinTable:
            bt = time.time()
            plot_name = path.join(dirPath, basinName + r".html")
            basin = parse_basin_name(basinName).upper()
            chart_title = f"{SENSOR_LONG[sensor]} in {basin}".upper()
            try:
                basinSites = getBasinSites(basinName, basinTable)
                fig_data, anno_text = update_chart(basinName, basinSites, all_meta, args)
                if not fig_data:
                    fig_data, anno_text = get_no_chart_fig_data(sensor)

                export_fig_data(fig_data, plot_name)
                fig = go.Figure(fig_data)
                if args.dev:
                    save_dev_plot(fig, plot_name, anno_text, chart_title)
                else:
                    save_prod_plot(fig, plot_name, anno_text, chart_title)
                charts_count += 1
                chart_pub_str = (
                    f"     {basinName} {script_name_str} "
                    f"Chart created in "
                    f"{round(time.time()-bt,2)} seconds"
                )
                print(chart_pub_str)

            except Exception as e:
                exc_type, exc_value, exc_traceback = sys.exc_info()
                error_trace = traceback.format_tb(exc_traceback)
                error_line = error_trace[-1].split(",")[1].strip()
                e = str(e).replace(",", " ")
                err_str = (
                    f"     No {script_name_str} "
                    f"chart created for {basinName} "
                    f"- {e} - {error_line}"
                )
                print(err_str)

                logging_str = (
                    f"CHART FAILED,"
                    f'{scriptName.replace("_",",")},'
                    f'{state},{basinName.replace(","," ")},{e},'
                    f'{error_line.replace("line ","")},'
                    f'{today.strftime("%Y-%m-%d,%H:%M:%S")}'
                )

                logging.error(logging_str)
                err_count += 1

    logging_str = (
        f"# CHARTS PRODUCED,"
        f"{charts_count},"
        f'{scriptName.replace("_",",")},'
        f'{today.strftime("%Y-%m-%d,%H:%M:%S")}'
    )
    logging.info(logging_str)
    logging_str = (
        f"# CHARTS FAILED,"
        f"{err_count},"
        f'{scriptName.replace("_",",")},'
        f'{today.strftime("%Y-%m-%d,%H:%M:%S")}'
    )
    logging.info(logging_str)
```
